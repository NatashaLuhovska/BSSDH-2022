## Filtering Documents with Pandas

```python
import pandas
```

```python
 
 url_1 = "https://github.com/ValRCS/BSSDH_22/raw/main/corpora/ua_old_newspapers.zip"

 df_1 = pandas.read_csv(url_1, sep="\t", compression="zip")
 df_1.head(15)
```

```python
# the size of the corpus:
print(len(df_1))
```

```python
df_1.shape
```

```python
# show the last 20 entries

df_1.tail(20)
```

```python
df_1["Language"].unique()
```

```python
df_1.Source.unique()
```

```python
unian_df = df_1[df_1.Source == 'unian.net']
unian_df.shape
```

```python
unian_df.head()
```

```python
document_list = list(unian_df.Text)
len(document_list)
```

```python
document_list[:10]
```

```python
unian_Kyiv = unian_df[unian_df.Text.str.contains("Київ")]
unian_Kyiv.shape
```

```python
unian_Kyiv_100 = unian_Kyiv.sample(100) # could use head(100) or tail(100)
# for full credit do not just use this single command I want to see some filtering beforehand
unian_Kyiv_100.shape
```

```python
unian_Kyiv_100.to_csv("unian_Kyiv_100.csv", sep="\t")
```
